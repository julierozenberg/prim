```python
from itertools import accumulate
def plot_anova(varin,nameofvarin,ax):
    #formula=varin+" ~ agchange+manuchange+empchange+grag+grmanu+grserv+skillpserv+skillpag+skillpmanu+b+isssp5"
    formula=varin+" ~ agchange+manuchange+empchange+grserv+skillpserv+grag+skillpag+grmanu+skillpmanu+b+p+isssp5"
    olsmodel=ols(formula,data=concat([outcomes,experiments],axis=1)).fit()
    table=anova_lm(olsmodel)
    table.index=["Change\nin the share\nof agriculture","Change\nin the share\nof industry","Change\nin participation",
                 "Growth for\nservice workers\n(unskilled)","Skill premium\n(services)",
                 "Growth for\nagriculture workers\n(unskilled)","Skill premium\n(agriculture)",
                 "Growth for\nindustry workers\n(unskilled)", "Skill premium\n(manufacture)",
                 "Basic income", "Redistribution\nto the elderly","Population","Combination\nof parameters\n(residual)"]
    table['sum_sq_pc']=table['sum_sq']/table['sum_sq'].sum()
    table['colors']=["#8dd3c7","#ffffb3","#bebada","#fb8072","#80b1d3","#fdb462","#b3de69","#fccde5","#d9d9d9",
                      "#bc80bd","#ccebc5","#ffed6f","#b15928"]
    table=table.sort(['sum_sq'],ascending=False)
    cumu = list(accumulate(table['sum_sq_pc']))
    x    = table['sum_sq_pc']
    prev = 0
    for i,j in enumerate(x):
        if cumu[i]<0.9:
            ax.plot([0,0],[prev,prev+j],color="black")
            ax.plot([1,1],[prev,prev+j],color=table.loc[x.index[i],'colors'])
            ax.fill_between([0,1], prev, prev+j,facecolor=table.loc[x.index[i],'colors'],alpha=0.6)
            ax.annotate(x.index[i], xy=(0.05,np.mean([prev,prev+j])-0.01),  xycoords='data',
                            xytext=(0,0), textcoords='offset points'
                                )
            prev += j
        else:
            ax.plot([0,0],[prev,1],color="black")
            ax.plot([1,1],[prev,1],color="black")
            ax.fill_between([0,1], prev, 1,facecolor="grey",alpha=0.6)
            ax.annotate("Others", xy=(0.05,np.mean([prev,1])),  xycoords='data',
                            xytext=(0,0), textcoords='offset points'
                                )
            break

    ax.spines['right'].set_visible(False)
    #ax.spines['left'].set_visible(False)
    ax.xaxis.set_ticks([])
    ax.spines['bottom'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.set_ylim([0,1])
    ax.set_xlim([0,1])
    ax.set_ylabel("{}, % of variance explained by drivers".format(nameofvarin))
```

```python
font = {'family' : 'Bitstream Vera Sans',
    'weight' : 'normal',
    'size'   : 10}
plt.rc('font', **font)
f, axarr = plt.subplots(1,4,figsize=(10,10))
plot_anova(varpoors,"Average income of the bottom 20%",axarr[0])
plot_anova("below6pc","Share of people below 6USD per day",axarr[1])
plot_anova("below10pc","Share of people below 10USD per day",axarr[2])
plot_anova(varineq,"Income share of the bottom 40%",axarr[3])
f.tight_layout()
plt.savefig("anova_different_groups_vnm.png",bbox_inches="tight",dpi=500)
```

```python
def plot2data(data1,data2,logical,thetitle="",yy=1.03):
    plt.plot(outcomes[data1],outcomes[data2],marker='o',linestyle='None',markersize=6,color='white')
    corrcoeff=round(np.corrcoef(outcomes[data1],outcomes[data2])[0,1],2)
    plt.text(outcomes[data1].min(),outcomes[data2].max(),'Correlation: {}'.format(corrcoeff))
    plt.title(thetitle,y=yy)
    #plt.plot(outcomes.ix[logical,data1],outcomes.ix[logical,data2],marker='s',linestyle='None',markersize=8,color='red')
```

```python
def plotgroup(data1,data2,group,thelabel,thecolor,pos=0):
    plt.plot(outcomes.ix[group,data1],outcomes.ix[group,data2],marker='x',linestyle='None',markersize=6,color=thecolor)
    if pos==0:
        (xpos,ypos)=(100, 60)
        ang='40'
    elif pos==1:
        (xpos,ypos)=(-150, 130)
        ang='-40'
    elif pos==2:
        (xpos,ypos)=(-130, -140)
        ang='80'
    plt.annotate(thelabel, xy=(np.mean(outcomes.ix[group,data1]), np.mean(outcomes.ix[group,data2])),  xycoords='data',
                    xytext=(xpos,ypos), textcoords='offset points',
                    size=15,
                    #bbox=dict(boxstyle="round", fc="0.8"),
                    arrowprops=dict(arrowstyle="fancy",
                                    fc=thecolor, ec=thecolor,
                                    patchB=None,alpha=0.4,
                                    connectionstyle="angle3,angleA=0,angleB={}".format(ang)),
                        )
    
```

```python
def plotini(var1,var2):
    xini=ini_data.ix[ini_data['countrycode']==cc,var1]
    yini=ini_data.ix[ini_data['countrycode']==cc,var2]
    plt.plot(xini,yini,color="black",marker='x',markeredgewidth=1,linestyle='None',markersize=12,clip_on=False)

    plt.annotate('today', xy=(xini, yini),  xycoords='data',
                    xytext=(-70, 20), textcoords='offset points',
                    size=15,
                    #bbox=dict(boxstyle="round", fc="0.8"),
                    arrowprops=dict(arrowstyle="fancy",
                                    fc="0.6", ec="none",
                                    patchB=None,
                                    connectionstyle="angle3,angleA=0,angleB=30"),
                        )
```
