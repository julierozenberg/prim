```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'forprim_MWI_2030.csv')
df['isssp5']=(df.ix[:,'ssp']=='ssp5')+0
columns = ['isssp5','agchange','manuchange','empchange','gr1','gr2','gr3','gr4','gr5','gr6','b','p']
x = df.ix[:,columns]
y = df.ix[:,'opt'].values
import numpy.lib.recfunctions as recfunctions

x = x.to_records()
x = recfunctions.drop_fields(x, 'index')
```

fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(y, bins=20)
plt.show()

```python
from expWorkbench import ema_logging
from analysis import prim

ema_logging.log_to_stderr(ema_logging.INFO)

def classify(outcomes):
    outcome = outcomes['y']
    classes = np.zeros(outcome.shape)
    
    classes[(outcome==1)] =1
    return classes
    

prim_obj = prim.Prim((x,{'y':y}), classify, threshold=.6)
#prim_obj.perform_pca()
```

```python
box1 = prim_obj.find_box()
```

```python
box1.show_tradeoff()
plt.show()
box1.show_ppt()
plt.show()
```

```python
box1.inspect(31)
```

```python
# just to get the last box, any integer can be used
index_last_box = box1._cur_box

box_lim = box1.box_lims[index_last_box]
res=pd.DataFrame.from_records(box_lim)
```

```python
res
```

```python
select=""
for thecol in res.columns:
    select+="(df['{}']>={})&(df['{}']<={})&".format(thecol,res.ix[0,thecol],thecol,res.ix[1,thecol])
select=select[:-1]
```

```python
y2=y.copy()
y2[eval(select).values]=0
```
