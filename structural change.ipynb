```python
from pandas.io import wb
import numpy as np
from pandas import DataFrame,read_csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
groups = read_csv("income_groups.csv",header =4,index_col=2)
groups= groups.loc[:,["Code","Region","Income group","Lending category","Other"]]#,
groups.dropna(axis=0,how="all");
groups.rename(columns={"Other":"HIPC"},inplace=True)
```

```python
def get_wb_av(wb_name):
    """calculates average growth rate on """
    #return all values
    wb_raw =(wb.download(indicator=wb_name,start=1960,end=2013,country="all"))
    #removes nans, and takes the most revent value. hop has a horrible shape
    hop=wb_raw.reset_index().dropna().groupby("country").apply(lambda x:f(x,wb_name))
    hop= hop.dropna().reset_index().drop("level_1",axis=1)
    return hop

def f(x,wb_name):
    out = DataFrame(columns=['ini_year','end_year','ini_value','lastvalue'])
    out['ini_year'] = x['year']
    out['end_year'] = x['year'].astype(int)+20
    for ini_year in x['year'].unique():
        if (int(ini_year)<1994)&(sum(x['year']==str(int(ini_year)+20))>0):
            beg=x.ix[x['year']==ini_year,wb_name]
            end=x.ix[x['year']==str(int(ini_year)+20),wb_name]
            out.ix[out['ini_year']==ini_year,'ini_value'] = beg.values 
            out.ix[out['ini_year']==ini_year,'lastvalue'] = end.values 
    return out
```

```python
ag_emp = get_wb_av('SL.AGR.EMPL.ZS')
```

```python
indus_emp = get_wb_av('SL.IND.EMPL.ZS')
```

```python
emp_ratio = get_wb_av('SL.EMP.TOTL.SP.NE.ZS')
```

```python
def new_df(dfin):
    out = DataFrame()
    out['inirounded'] = dfin['inirounded'].unique()
    for ini in dfin['inirounded'].unique():
        raw = out['inirounded']==ini
        out.ix[raw,'min'] = 1/100*dfin.ix[dfin['inirounded']==ini,'lastvalue'].apply(np.round).min()
        out.ix[raw,'max'] = 1/100*dfin.ix[dfin['inirounded']==ini,'lastvalue'].apply(np.round).max()
    
    out = out.sort('inirounded')
    return out
```

```python
ag_emp['inirounded'] = 1/100*ag_emp.ini_value.apply(np.round)
indus_emp['inirounded'] = 1/100*indus_emp.ini_value.apply(np.round)
emp_ratio['inirounded'] = 1/100*emp_ratio.ini_value.apply(np.round)
```

```python
outag = new_df(ag_emp)
outind = new_df(indus_emp)
outemp = new_df(emp_ratio)
```

```python
def fit_kernel(out):
    from statsmodels.nonparametric.kernel_regression import KernelReg
    x = out['inirounded']
    y1 = out['min']
    y2 = out['max']
    kr1 = KernelReg(y1,x,'c')
    kr2 = KernelReg(y2,x,'c')
    #plt.plot(x, y, '+')
    y_pred1, y_std1 = kr1.fit(x)
    y_pred2, y_std2 = kr2.fit(x)
    plt.plot(x, y_pred1,'o',label='Min historical data')
    plt.plot(x, y_pred2,'o',label='Max historical data')
    plt.plot([0,1],[0,1])
    return y_pred1,y_pred2
```

```python
from scipy.interpolate import UnivariateSpline
x    = [0,0.01,0.1,0.3,0.5,0.7,0.9,1]
ymin = [0,0.001,0.01,0.1,0.2,0.3,0.4,0.6]
yminemp  = [0,0.007,0.07,0.25,0.4,0.6,0.75,0.8]
ymax      = [0,0.11,0.3,0.6,0.7,0.9,0.99,1]
w         = [1,2,2,2,1,1,1,1]
smin      = UnivariateSpline(x, ymin, w)
sminemp      = UnivariateSpline(x, yminemp, w)
smax      = UnivariateSpline(x, ymax, w)

plt.plot(x,smin(x))
plt.plot(x,smax(x))
plt.plot([0,1],[0,1])
```

```python
agmin,agmax=fit_kernel(outag)
x    = [0,0.01,0.1,0.3,0.5,0.7,0.9,1]
yminag = [0,0.001,0.01,0.1,0.2,0.3,0.4,0.6]
ymaxag      = [0,0.2,0.15,0.4,0.5,0.6,0.8,0.8]
w         = [1,2,2,2,1,1,1,1]
sminag      = UnivariateSpline(x, yminag, w)
smaxag      = UnivariateSpline(x, ymaxag, w)
plt.plot(x,sminag(x),'blue',label='Min uncertain range')
plt.plot(x,smaxag(x),'green',label='Max uncertain range')
plt.xlabel("Initial share of agriculture (% of employment)")
plt.ylabel("Share of agriculture 20 years later")
plt.legend(loc='upper left',frameon=False,numpoints=1)

plt.savefig('ag_struct_change.png',bbox_inches="tight",dpi=300)
```
