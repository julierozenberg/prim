```python
def sumsoverallcountries(df):
    out=df[['quintile2', 'below1.25', 'below4','tot_pop', 'GDP']].sum()
    out['poors']=out['below1.25']/out['tot_pop']
    out['nearpoors']=out['below4']/out['tot_pop']
    out['quintilepc2']=out['quintile2']/out['GDP']
    return out
```

```python
def worldgdpgrowth(ssp,year,countrylist):
    sspstring='SSP{}_v9_130325'.format(ssp)
    gdp_2007=0
    gdp_fin=0
    r32_list=list()
    for cc in countrylist:
        r32=country2r32(codes_tables,cc)
        r32_list.append(r32)
    r32_set=set(r32_list)
    for reg in r32_set:
        selection=(ssp_gdp['REGION']==reg)&(ssp_gdp['MODEL']=='OECD Env-Growth')&(ssp_gdp['SCENARIO']==sspstring)
        y1=ssp_gdp.ix[selection,'2005'].values[0]
        y2=ssp_gdp.ix[selection,'2010'].values[0]
        f=interpolate.interp1d([2005,2010], [y1,y2],kind='slinear')
        gdp_2007+=f(2007)
        gdp_fin+=ssp_gdp.ix[selection,str(year)].values[0]
    gdp_growth=gdp_fin/gdp_2007
    return gdp_growth
```

```python
def compare2ini(df,allini):
    compare=df[['shareag','sharemanu','shareemp']]-100*allini[['shareag','sharemanu','shareemp']]
    return compare
```

```python
def reindexdf(df):
    df=df.sort('countrycode')
    df.index = df['countrycode']
    return df
```

```python
from scipy.stats import gaussian_kde
def plot_distrib(y,thecolor,thetitle,leg):
    hist, bin_edges = np.histogram(y,bins=100,density=True)
    width = 0.5 * (bin_edges[1] - bin_edges[0])
    center = (bin_edges[:-1] + bin_edges[1:]) / 2
    xs = np.linspace(center.min(),center.max(),300)
    #plt.bar(center, hist, align='center', width=width,alpha=0.05,color=thecolor)
    density = gaussian_kde(y)
    density._compute_covariance()
    plt.plot(xs,density(xs),linewidth=2,color=thecolor,label=leg)
    plt.title(thetitle)
```

```python
plt.figure(figsize=(12,10))
i=1
thecolor=["#9900FF","#CC00FF","#CC33FF","#CC66FF","#CC99FF","#CCCCFF","#CCFFFF"]
thecolor2=["#330000","#333300","#336600","#339900","#33CC00","#33FF00"]
titles=["Growth services unskilled","Growth services skilled","Growth ag unskilled","Growth ag skilled",
        "Growth manu unskilled","Growth manu skilled"]
for var in ['gr1','gr2','gr3','gr4','gr5','gr6']:
    plt.subplot(3,2,i)
    plot_distrib(allopt[var],thecolor2[4],titles[i-1],"optimistic scenario")
    plot_distrib(allpess[var],thecolor[4],titles[i-1],"pessimistic scenario")
    if i%2==0:
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    i+=1
```
